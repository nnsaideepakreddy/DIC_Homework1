### Problem Statement and Feature Selection
Problem Statement: We want to conduct an analysis of the trends of vehicle thefts across Los Angeles by location, time of day, and victim demographics. Through a focus on these variables, we can identify high-risk locations and times and even possibly identify targeted groups.

Most Important Features:Â 

* Date - DATE OCC and Time - TIME OCC: knowing when
* Location - AREA NAME, LAT, LON: knowing where.

The demographics to describe the victim are Vict Age, Vict Sex, Vict Descent. These are added into the analysis to see if there are specific demographics that are targeted more than others. 

Crime Code Description: Crm Cd Desc filters for vehicle-related thefts.

Premises: Premis Desc describes the surroundings of the crime.

### Approach to the Problem Statement
Data Filtering: We will filter for only vehicle theft incidents: Crm Cd Desc contains "VEHICLE - STOLEN".

Data Exploration: We perform exploratory data analysis to identify possible relationships between thefts with time, location, and victim demographics using visualization.

Hypothesis Testing: We can look, for instance, at whether vehicle theft is more common in downtown areas and/or at nighttime according to our visual investigation.

### Data Cleaning Steps
Filter Relevant Data: Filter out all rows which do not pertain to vehicle theft.

Handle Missing Data: Drop rows with missing values in essential fields or fill appropriately.

Convert DATE OCC and TIME OCC to datetime: Convert these columns to enable time-based analysis.

Remove Unwanted Features

DR_NO: Identifier field, little use in terms of analytics.
Date Rptd: More interested in when the occurrence actually occurred.

Rpt Dist No, Weapon Used Cd, Weapon Desc, etc.: Those do not add anything to the analysis of vehicle thefts.

Status, Crm Cd 2, Crm Cd 3, Crm Cd 4, etc.: Not needed for this investigation.

```python
df_vehicle_thefts = df[df['Crm Cd Desc'] == 'VEHICLE - STOLEN']
```

```python
df_vehicle_thefts_clean = df_vehicle_thefts.dropna(subset=['LAT', 'LON', 'DATE OCC', 'TIME OCC'])
```

```python
df_vehicle_thefts_clean['DATE OCC'] = pd.to_datetime(df_vehicle_thefts_clean['DATE OCC'])
df_vehicle_thefts_clean['TIME OCC'] = df_vehicle_thefts_clean['TIME OCC'].apply(lambda x: pd.to_datetime(str(x).zfill(4), format='%H%M').time())
```

```python
df_vehicle_thefts_clean['Day of Week'] = df_vehicle_thefts_clean['DATE OCC'].dt.day_name()
df_vehicle_thefts_clean['Hour'] = df_vehicle_thefts_clean['TIME OCC'].apply(lambda x: x.hour)
```

```python
columns_to_drop = ['DR_NO', 'Date Rptd', 'Rpt Dist No', 'Part 1-2', 'Mocodes', 
                   'Premis Cd', 'Weapon Used Cd', 'Weapon Desc', 'Status', 
                   'Crm Cd 1', 'Crm Cd 2', 'Crm Cd 3', 'Crm Cd 4', 
                   'Cross Street', 'Premis Cd']

df_vehicle_thefts_clean = df_vehicle_thefts_clean.drop(columns=columns_to_drop)
df_vehicle_thefts_clean.info()
```

### Hypothesis and justification:
Based on the given visualizations, here are some hypotheses:

Geographic distribution of vehicle thefts:
Hypothesis: Vehicle thefts do not occur uniformly in areas but rather there are pockets within a neighborhood that have extremely high rates of theft.

Evidence: In the "Vehicle Thefts by Area" bar chart, there's quite a variation within different areas of theft counts, ranging from over 60,000 thefts in the "Central" area to about 25,000 thefts in "Foothill".

Patterns in age and gender of victims
Hypothesis: Vehicle theft victimization does vary with age and gender, and maybe males might get targeted more frequently over a wider range of ages.

Supporting evidence: The box plot "Victim Age Distribution by Gender" appears to show a wider interquartile range for the males (M), as compared to the females (F). This displays greater variability in the age distribution of the male victims.

Patterns in vehicle thefts over the week:
Hypothesis: More vehicle thefts occur during the week, especially mid-week, than on weekends.

Evidence: The "Vehicle Thefts by Day of the Week" chart indicates that Thursday has the highest number of vehicle thefts, with Wednesday and Tuesday following closely behind. Saturday and Sunday represent the smallest numbers.

Time of day in which vehicle thefts occur:
Hypothesis: Most occurrences of vehicle theft fall in the evening and nighttime hours.

Evidence: In the graph "Vehicle Thefts by Hour of the Day", there is a tendency for the counts of thefts to go up in the evening-after 9 PM-21:00-and to be relatively high through early morning hours, and the lowest counts are recorded during early morning, around 5-6 AM.

Urban vs. rural theft rates:
Hypothesis: The higher the population or more urban in character an area is, the higher the rate of vehicle thefts.

Evidence: As perhaps somewhat implicitly shown by the "Central" high theft count from the first chart, the more densely populated or central urban areas might be more prone to vehicle thefts.

Problem Statement: The goal is to study the trend and incidences of identity theft in Los Angeles so as to gain insight into the nature of places, points in time, and even victim demographics. This would serve useful for law enforcement strategies and also public awareness campaigns.

Approach:

Feature Selection: A few important features which give insight into the pattern of identity theft are:

Time of Occurrence (TIME OCC): It indicates at what time the incidence of identity theft is most likely to occur.

Date of Occurrence (DATE OCC): This allows for temporal trend analysis, whereby the variation in a selected crime type is watched over time, for example, a year, month, or season.

Area and Area Name (AREA, AREA NAME): This allows the determination of hotspots where identity theft occurs most.

Premises Code and Description (Premis Cd, Premis Desc): It lists what kind of premises the identity theft occurs within, such as open lots, residences, or commercial spaces.

The age and sex of the victim: For instance, identification of age groups or sex that might fall victim more easily to identity theft. Victims' demographic analysis can be done based on their age and sex.

Status Description: Current legal status of the case given, such as whether it is under investigation, an arrest has been made, etc.

Crime Code and Description: The details of different types of identity theft crimes can be obtained, such as online fraud or credit card theft.
